```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\NOBEL\Desktop\Python\securityupdate.csv')
df.head()
```

```python

gen_sui = df.pivot_table('Do you concern/know about your data security?', index='Gender', aggfunc='sum')
x = gen_sui.index.values
y = gen_sui.values
y = y.reshape(2,)

fig, ax = plt.subplots(figsize=(9,3))
explode = (0.1,0.1)
ax.pie(y, explode=explode, labels=x, autopct='%1.1f%%', shadow=True, startangle=0)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
